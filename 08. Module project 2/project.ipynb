## Проверка правильности вычисления эффективности обогащения

```python
from sklearn.metrics import mean_absolute_error
```

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']
```

```python
# Enrichment efficiency calculation
data_train['rougher.output.recovery_test'] = (C * (F - T)) / (F * (C - T)) * 100
```

### Вспомогательные функции

```python
def get_nans_amount(feature_name):
    """
    Function for calculation of an amount of a missing values in some column

    Parameters:
    feature_name (string): Feature name
    """
    
    print('NaNs in {} column - {}'.format(feature_name, len(data_full[data_full[feature_name].isnull()])))
```

```python
def feature_distr_plot(feature_name, bins=100, show_small_values=False, small_values_threshold=1):
    """
    Function for plotting a histogram of the distribution of some feature and additional distributon for values,
    smaller than threshold

    Parameters:
    feature_name (string): Feature name
    bins (int): Amount of bins on histogram, default=100
    show_small_values (bool): Flag for additional distributon plot, default=False
    small_values_threshold (float): Threshold for additional distributon, default=1
    """
    
    if show_small_values:
        fig, (ax1, ax2) = plt.subplots(
            nrows=2, ncols=1,
            figsize=(20, 10)
        )
    else:
        fig, ax1 = plt.subplots(
            nrows=1, ncols=1,
            figsize=(20, 10)
        )
    
    data_full[feature_name].hist(bins=bins, figsize=(15, 10), ax=ax1)
    ax1.set_xlabel(feature_name + ' value')
    ax1.set_ylabel('Number of objects')
    ax1.grid(linewidth=0.5)
    ax1.set_title('Objects distribution by ' + feature_name)
    
    if show_small_values:
        data_full[data_full[feature_name] <= small_values_threshold][feature_name].hist(bins=bins, figsize=(15, 10), ax=ax2)
        ax2.set_xlabel(feature_name + ' value')
        ax2.set_ylabel('Number of objects')
        ax2.grid(linewidth=0.5)
        ax2.set_title('Objects distribution by ' + feature_name + ' with values lower then or equals ' + str(small_values_threshold))
    
    plt.show()
```

#### Вспомогательные функции

```python
def show_plots_and_nans(feature_name, show_small_values=False, small_values_threshold=1):
    """
    Function that combines get_nans_amount() and feature_distr_plot() functions

    Parameters:
    feature_name (string): Feature name
    show_small_values (bool): Flag for additional distributon plot, default=False
    small_values_threshold (float): Threshold for additional distributon, default=1
    """
    
    get_nans_amount(feature_name)
    feature_distr_plot(feature_name, show_small_values=show_small_values, small_values_threshold=small_values_threshold)
```

```python
def replace_outliers_and_fill_nans(feature_name, threshold=0.01):
    """
    Function that:
    - replaces outliers (values lower than or equals selected threshold) with NaNs
    - filling NaNs with previos values (with limit = 1)
    - showing amount of NaNs and distribution of selected feature after previous operations.

    Parameters:
    feature_name (string): Feature name
    threshold (float): Threshold for replacing with NaNs, default=0.01
    """
    
    data_full.loc[data_full[feature_name] <= threshold, feature_name] = np.NaN
    
    data_full[feature_name].fillna(method='ffill', limit=1, inplace=True)
    
    show_plots_and_nans(feature_name)
```

## Функция для вычисления итоговой sMAPE

```python
def symmetric_mean_absolute_percentage_error(y_true, y_pred):
    """
    Function for calculting Symmetric Mean Absolute Percentage Error (sMAPE)
    
    Parameters:
    y_true (array-like): Real target values
    y_pred (array-like): Predicted traget values
    
    """
    
    smape = 0
    
    for i in range(len(y_true)):
        smape += abs(y_true[i] - y_pred[i]) / ((abs(y_true[i]) + abs(y_pred[i])) / 2)
    
    return smape / len(y_true) * 100
```

```python
def final_smape(y_pred_rougher, y_true_rougher, y_pred_final, y_true_final):
    """
    Function for calculting final sMAPE
    
    Parameters:
    y_pred_rougher (array-like): Predicted rougher recovery values
    y_true_rougher  (array-like): Real rougher recovery values
    y_pred_final (array-like): Predicted final recovery values
    y_true_final (array-like): Real final recovery values
    """
    
    return 0.25 * symmetric_mean_absolute_percentage_error(y_pred=y_pred_rougher, y_true=y_true_rougher) + 0.75 * symmetric_mean_absolute_percentage_error(y_pred=y_pred_final, y_true=y_true_final)
```
